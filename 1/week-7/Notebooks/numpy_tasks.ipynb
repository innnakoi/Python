```python
import numpy as np
```

```python
arr1 = np.arange(1, 11)
print(arr1)

arr2 = np.random.randint(1, 20, size=(3, 3))
print(arr2)
```

```python
print("Sum:", arr1.sum())
print("Mean:", arr1.mean())
print("Max:", arr1.max())

arr_multip = arr1 * 2
print("Multiplied by 2:", arr_multip)
arr_repl = np.where(arr1 > 5, 0, arr1)
print("Replaced values:", arr_repl)
```
