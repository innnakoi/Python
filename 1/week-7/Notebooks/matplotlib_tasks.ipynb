```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "Name": ["Man", "Max", "Charlie", "Diva", "Kimi"],
    "Age": [22, 27, 25, 30, 19],
    "Score": [77, 87, 78, 90, 87]
})
```

```python
names = df["Name"]
scores = df["Score"]

```

```python
plt.bar(names, scores)
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Bar Chart")
plt.show()
```

```python

plt.plot(names, scores)
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Line Plot")
plt.show()
```
